```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poly_leg(n):
    x = np.arange(-1,1,.01) #vector de -1 a 1
    
    coef = (1/2)**n #use lo que dice wiki
    
    Poly=0
    for k in range(n+1): #Formula de rodriguez
        pp =(np.math.factorial(n)/(np.math.factorial(k)*(np.math.factorial(n-k))))**2*(x+1)**(n-k)*(x-1)**k
        Poly = pp+Poly
    
    return coef*Poly #Regresa el polinomio de legendre en forma de arreglo
```

```python
#cmap = ListedColormap(['r', 'g', 'b'])
x = np.arange(-1,1,.01)
y=poly_leg(6)
plt.plot(x,poly_leg(0), x,poly_leg(1), x,poly_leg(2), x,poly_leg(3), x,poly_leg(4), x,poly_leg(5))
plt.xlim(-1.01,1.01)
plt.ylim(-1.01,1.05)
```
